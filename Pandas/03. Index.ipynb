```python
import pandas as pd

data={
    '이름': ['채치수','채진수','송태섭','서태풍','강백호','변덕규','황태산','윤대협'],
    '학교': ['북산고','북산고','북산고','북산고','북산고','능남고','능남고','능남고'],
    '키': [197, 184, 168, 187, 188, 202, 188, 190],
    '국어': [90,40,80,40,15,80,55,100],
    '영어': [85,35,75,60,20,100,65,85],
    '수학': [85,35,75,60,20,100,65,85],
    '과학': [85,35,75,60,20,100,65,85],
    '사회': [85,35,75,60,20,100,65,85],
    'SW특기': ['Python','Java','JS','','','C','Python','C#']
}

df=pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name='지원번호'
df
```

## Index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop=True) #원래 쓰던 '지원 번호' 인덱스 삭제 단 실제 데이터에는 반영되지 않는다.
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True) # 실제 데이터에 바로 반영
df
```
