```python
from collections import OrderedDict
```

```python
friend_ordered_list =OrderedDict([
    ('name', ['Jone','Nate']),
    ('age',[25,30]),
    ('jop', ['student','teacher'])
])
```

```python
df=pd.DataFrame.from_dict(friend_ordered_list)
```

```python
df.head()
```

```python
friend_list=[
    ['Jone',25,'student'],
    ['Nate',30,'teacher']
]
```

```python
column_name=['name','age','job']
```

```python
df=pd.DataFrame.from_records(friend_list,columns=column_name)
```

```python
df.head()
```
